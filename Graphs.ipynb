```python
from random import choice

class RandWalk():
    def __init__(self, num_points=5000):
        self.num_points = num_points
        
        self.x_val = [0]
        self.y_val = [0]
    
    def fill_walk(self):
        while len(self.x_val) < self.num_points:
            x_dir = choice([1, -1])
            x_dis = choice([0, 1, 2, 3, 4])
            x_step = x_dir * x_dis
        
            y_dir = choice([1, -1])
            y_dis = choice([0, 1, 2, 3, 4])
            y_step = y_dir * y_dis
            
            if x_step == 0 and y_step == 0:
                continue
            
            next_x = self.x_val[-1] + x_step
            next_y = self.y_val[-1] + y_step
        
            self.x_val.append(next_x)
            self.y_val.append(next_y)
```

```python
import matplotlib.pyplot as plt

rw = RandWalk()
rw.fill_walk()
plt.scatter(rw.x_val, rw.y_val, s=15)
plt.show()
```

```python
while True:
    rw = RandWalk()
    rw.fill_walk()
    point_num = list(range(rw.num_points))
    plt.scatter(rw.x_val, rw.y_val, c=point_num, cmap=plt.cm.Blues,
                edgecolor='none', s=15)
    plt.show()
    
    keep_run = input('Make another walk? (y/n): ')
    if keep_run == 'n':
        break
```

```python
class Die():
    def __init__(self, num_sides=6):
        self.num_sides = num_sides
    
    def roll(self):
        return randint(1, self.num_sides)
```

```python
from random import randint

die = Die()
results = []
for roll_num in range(1000):
    result = die.roll()
    results.append(result)
    
print(results)
```

```python
freq = []
for value in range(1, die.num_sides+1):
    freqy = results.count(value)
    freq.append(freqy)
    
print(freq)
```

```python
import pygal

for value in range(1, die.num_sides+1):
    freqy = results.count(value)
    freq.append(freqy)
    
hist = pygal.Bar()

hist.title='Results of rolling one D6 1000 times'
hist.x_labels = ['1', '2', '3', '4', '5', '6']
hist.x_title = "Result"
hist.y_title = 'Freq of Result'
hist.add('D6', freq)
hist.render_to_file('die_visual.svg')
```
